```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('nyc-taxi-weather-combined-2015.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop(columns=['city', 'date'], axis=1, inplace=True)
```

```python
df['zip_code']= df['zip_code'].astype('object')
```

```python
zip_code_group_df = df.groupby(['zip_code'])[['cnt']].sum().reset_index()
```

```python
zip_code_group_df
```

```python
## cacluating percentage from cnt numbers and round up at 3 decimal point

(zip_code_group_df[['cnt']]/df[['cnt']].sum()).round(3)
```

```python
zip_code_group_df['percent'] = (zip_code_group_df[['cnt']]/df[['cnt']].sum()).round(3).rename(columns={'cnt':'percent'}).reset_index()['percent']
```

```python
zip_code_group_df.head()
```

```python
zip_code_group_df.tail()
```

```python
## drop the rows where the percent 0 

zip_code_group_df = zip_code_group_df[zip_code_group_df['percent'] > 0.000]
```

```python
zip_code_group_df.tail()
```

```python
zip_code_group_df.sort_values(by='cnt', ascending=False)
```

```python
zip_code_group_df[zip_code_group_df['percent'] > 0.05]
```

```python
df['zip_code'] = df['zip_code'].astype(int)
```

```python
df.info()
```

```python
df['pickup_hour'] = pd.to_datetime(df['pickup_hour'])
```

```python
df['zip_code'] = df['zip_code'].astype(object)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['month'] = df['pickup_hour'].dt.month
df['day'] = df['pickup_hour'].dt.day
```

```python
df.head()
```

```python
df.tail()
```

```python
to_be_labeled_feature_mask = df.dtypes==object
categorical_cols = df.columns[to_be_labeled_feature_mask].tolist()
```

```python
le = LabelEncoder()
df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
```
